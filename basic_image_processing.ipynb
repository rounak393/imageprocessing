```python
# OpenCV

# Step 1: Install OpenCV
!pip install opencv-python-headless matplotlib

# Step 2: Import Required Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Utility function to display images using matplotlib
def show_image(img, title="Image", cmap_type='gray'):
    plt.figure(figsize=(6, 6))
    if len(img.shape) == 2:
        plt.imshow(img, cmap=cmap_type)
    else:
        img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
        plt.imshow(img)
    plt.title(title)
    plt.axis('off')
    plt.show()

# Step 3: Load an Image
# Upload a sample image using Colab's file upload
def upload_image():
    from google.colab import files
    uploaded = files.upload()
    for fn in uploaded.keys():
        img = cv2.imread(fn)
        show_image(img, title="Original Image")
        return img

# Step 4: Image Basics
# 4.1 Convert to Grayscale
# 4.2 Resize Image
# 4.3 Crop Image

def basic_operations(img):
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
    resized = cv2.resize(img, (200, 200))
    cropped = img[50:200, 100:250]  # Crop a region

    show_image(gray, title="Grayscale")
    show_image(resized, title="Resized")
    show_image(cropped, title="Cropped")
    return gray, resized, cropped

# Step 5: Drawing and Writing
# 5.1 Draw Shapes (Line, Rectangle, Circle)
# 5.2 Write Text on Image

def drawing_text():
    blank = np.zeros((300, 300, 3), dtype=np.uint8)
    cv2.line(blank, (0, 0), (300, 300), (255, 0, 0), 2)
    cv2.rectangle(blank, (50, 50), (250, 250), (0, 255, 0), 3)
    cv2.circle(blank, (150, 150), 50, (0, 0, 255), -1)
    cv2.putText(blank, 'OpenCV', (70, 280), cv2.FONT_HERSHEY_SIMPLEX, 0.8, (255, 255, 255), 2)
    show_image(blank, title="Drawing and Text")

# Step 6: Image Filtering
# 6.1 Blurring (Averaging, Gaussian, Median)
# 6.2 Edge Detection (Sobel, Canny)

def image_filtering(img):
    blur = cv2.GaussianBlur(img, (7, 7), 0)
    edges = cv2.Canny(img, 100, 200)
    show_image(blur, title="Gaussian Blur")
    show_image(edges, title="Canny Edges")

# Step 7: Thresholding
# 7.1 Simple Thresholding
# 7.2 Adaptive Thresholding

def thresholding(gray):
    _, th1 = cv2.threshold(gray, 127, 255, cv2.THRESH_BINARY)
    th2 = cv2.adaptiveThreshold(gray, 255, cv2.ADAPTIVE_THRESH_GAUSSIAN_C,
                                cv2.THRESH_BINARY, 11, 2)
    show_image(th1, title="Simple Threshold")
    show_image(th2, title="Adaptive Threshold")

# Step 8: Contours and Shape Detection

def contours_detection(gray):
    edges = cv2.Canny(gray, 50, 150)
    contours, _ = cv2.findContours(edges, cv2.RETR_EXTERNAL, cv2.CHAIN_APPROX_SIMPLE)
    contour_img = np.copy(cv2.cvtColor(gray, cv2.COLOR_GRAY2BGR))
    cv2.drawContours(contour_img, contours, -1, (0, 255, 0), 2)
    show_image(contour_img, title="Contours")

# Step 9: Morphological Operations
# 9.1 Dilation, Erosion, Opening, Closing

def morphology_operations(gray):
    kernel = np.ones((5, 5), np.uint8)
    dilated = cv2.dilate(gray, kernel, iterations=1)
    eroded = cv2.erode(gray, kernel, iterations=1)
    show_image(dilated, title="Dilated")
    show_image(eroded, title="Eroded")

# Step 10: Advanced Topics (Mathematical Background included in explanations)
# 10.1 Fourier Transform for Frequency Analysis
# 10.2 Histogram Equalization
# 10.3 Image Gradients (Sobel, Laplacian)

def advanced_processing(gray):
    # Fourier Transform
    f = np.fft.fft2(gray)
    fshift = np.fft.fftshift(f)
    magnitude_spectrum = 20 * np.log(np.abs(fshift))
    show_image(magnitude_spectrum, title="Fourier Magnitude Spectrum")

    # Histogram Equalization
    hist_eq = cv2.equalizeHist(gray)
    show_image(hist_eq, title="Histogram Equalized")

    # Sobel Gradients
    sobelx = cv2.Sobel(gray, cv2.CV_64F, 1, 0, ksize=5)
    sobely = cv2.Sobel(gray, cv2.CV_64F, 0, 1, ksize=5)
    magnitude = np.sqrt(sobelx**2 + sobely**2)
    show_image(magnitude, title="Sobel Gradient Magnitude")


```
